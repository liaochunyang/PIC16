# Exercise 1: Class of Cylinders

1. Create a class called **cylinder**. Initialization contains two arguments (**radius**, and **height**), which are provided by users. Each instance has two attribute (radius, and height) and values are assigned in initialization.

2. Define a magic method to compare two cylinders using `==`. Two cylinders are the same if both their radius and their heights are the same.

3. Define a method called volume to compute the volume for each cylinder. The mathematical formula is $\pi*r^2*h$ where $r$ is radius and $h$ is height. The final result is saved as another attribute (volume) for each instance.

4. Define a method called area to compute the area for cylinder. The mathematical formula is $2\pi r(r+h)$. The final result is saved as another attribute (area) for each instance.

5. Create another class called **cone**. This class is inherited from cylinder class. When you define its initialization for class cone, you call the initialization in class cylinder to get instance variable radius and height. You also need to create an attribute (named slant) for cone only. In other words, 3 arguments should be provided when you create a cone.

6. Define a method called volume to compute the volume for each cone. Inside this method, please call method volume in the cylinder class to get a number first and then divided by 3 to get the volume of a cone. The mathematical formula for the volume of a cone is $\frac{1}{3}\pi*r^2*h$ where $r$ is radius and $h$ is height. The final result is saved as another attribute (volume) for each instance.

7. Define a method called area to compute the area of each cone. The mathematical formula is $\pi r^2 + \pi*L*r$ where $r$ is the radius and $L$ is the slant height (attribute slant). The final result is saved as another attribute (area) for each instance.


```python
# Your solution here
class cylinder:
    
    def __init__(self,r,h):
        self.radius = r
        self.height = h
        
    def __eq__(self,other):
        
        return self.radius == other.radius and self.height == other.height
    
    def volume(self):
        
        self.volume = pi*self.radius**2*self.height
        return self.volume
        
    def area(self):
        
        self.area = 2*pi*self.radius*(self.radius+self.height)
        return self.area
        
class cone(cylinder):
    
    def __init__(self,r,h,L):
        
        super().__init__(r,h)
        self.slant = L
        
    def volume(self):
        
        self.volume = super().volume()/3
        return self.volume
        
    def area(self):
        self.area = pi*self.radius**2 + pi*self.slant*self.radius
        return self.area

```

```python
# Test example 1:
cylinder1 = cylinder(1,1)
cylinder2 = cylinder(1,2)

print(cylinder1 == cylinder2)    # should print False

print(cylinder1.volume())        # should print pi

print(cylinder2.volume())        # should print 2*pi

print(cylinder1.area())          # should print 4*pi
```

```python
# Test example 2:
cone1 = cone(1,1,1)
print(cone1.volume())          # should print pi/3
print(cone1.area())            # should print 2*pi
```
