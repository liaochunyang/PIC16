# Function

Function is a block of code which runs when it is called. 

Usually, it accepts passing inputs (can be empty) and returns outputs (None if it is empty). 


```python
def A( (B) ):
    """
    Docstring.
    """
    (C)

````

- A: function name. it is informative. 
- (B): passing inputs/arguments/inputs. We can have multiple inputs or empty input. Different inputs are separated by comma.
- (C): function body.
- (D): it is your (multiple) outputs, separated by comma. Return is optional, it will return None if there is no return.
- Docstring: explanation/comments of your function. It is a userbook. 

`````python
# Example: compute absolute value
def abs_val(n):
    
    """
    Compute the absolute value. 
    Args:
    n: It is a number (int or float number). 
    Returns:
    d: The absolute value of n. It has the same data type as n. 
    """
    
    if n>0:
        d = n
        return d
    else:
        d = -n
        return d
`````

`````python
# Example: Given a list of numbers, then return the absolute value of each number and save them as a list. 
# (At least three ways to do it. Different loops. Different ways to iterate.)

def abs_val_list(L):
    """
    Compute the absolute value.
    Args:
    L: It is a list of numbers. L is a list. L only contains int or float.
    Returns:
    M: List. Absolute values of elements in L.
    """
    M = [ ]
    
    for i in L:
        
        if i>0:
            M.append(i)
        else:
            M.append(-i)
            
    return M
`````

`````python
def abs_val_list(L):
    """
    Compute the absolute value.
    Args:
    L: It is a list of numbers. L is a list. L only contains int or float.
    Returns:
    M: List. Absolute values of elements in L.
    """
    
    for idx in range(len(L)):     # Method 3: Here we can use enumerate. 
        
        i = L[idx]
        
        if i<0:
            L[idx] = -i
            
    return L
`````

# Function Scope

Def (local variable/scope): A python variable is a local variable if it is defined and limited inside a function. 

Def (global variable): is a python variable that is globally accessible. Usually, it is defined outside a function. 

Rule: A variable is assigned to during the function call, then it is local to this function.  -- Oct 9

`````python
# Exercise: List global and local variables.

L = [1,2,3,4]          # global   
s = 10                 # global      

def sum_of_list(M):    # local
    """
    Compute the sum of elements in a list.
    Args:
    M: List of numbers.
    Returns:
    s: sum of all elememts.
    """
    s = 0              # local 
    for i in M:
        s = s + i
        
    return s
    
`````

# Variable number of arguments

Motivation: It is not efficient if we fix the number of passing arguments. Python does allow variable numbers of passing arguments.

**Example: Range function** range(4), range(1,10), range(1,10,2)

Goal: Write a function which allows variable numbers of passing arguments.

Keywords:

- *args: 1) passing arguments are saved as a tuple, 2) we use positional arguments to call the function.

- \*\*kwargs: passing arguments are saved as a dictionary, 2) keyword arguments are used.

Purpose: Keywords for variable numbers of passing arguments.


`````python
# Not efficient if we fix the number of passing arguments

def sum_of_two(x,y):
    return x+y

def sum_of_three(x,y,z):
    return x+y+z
`````

`````python
def sum_of_numbers(*args):
    
    s = 0    # total
    for i in args:
        s = s + i
        
    return s

# *args means positional arguments, when we call the function, we use positional arguments
print( sum_of_numbers(1,2,6,10,100) )
print( sum_of_numbers(1,2) )
`````

`````python
def sum_of_numbers_keyword(**kwargs):
    
    s = 0     # total
    
    #for key in kwargs:
    #    value = kwargs[key]
        
    
    for key, value in kwargs.items():
        s = s + value
    return s

print(  sum_of_numbers_keyword(x=1, y=2, z=3)  )
print(  sum_of_numbers_keyword(a=1, b=2, c=3)  )
`````
