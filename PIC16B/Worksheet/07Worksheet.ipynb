# Worksheet: Auto-differentiation

The goal of this worksheet is to use tensorflow and torch to compute the gradient.

## Warm-up

Your TA will give several warm-up examples.

## Exercises:

1. Compute the gradient of $f(x) = \sqrt{\|x\|_2} $ at $x=(0,0,0,0)$.

2. Compute the gradient of $g(x) = \frac{1}{2+\cos(x_1+x_2+x_3)}$ at $x^{(1)}=(1,0,1)$ and $x^{(2)}=(0,0,0)$.

For each exercise, you can use either tensorflow or torch.

**Grading is based on completion.**
