# Oct 17 Magic Method

Magic methods are defined with double underscores at the beginning and the end, i.e. \_\_init\_\_. We can use magic method to overload built-in operators, i.e. "+", "*", "==", "print", and etc.

````python
class MyList:
    
    def __init__(self, L):
        """
        Initialization.
        Args:
        L: built-in Python List.
        """
        self.list = L
        
    def __add__(self, other):
        """
        Element-wise addition.
        """
        # check lengths
        len1 = len( self.list )
        len2 = len( other.list )
        
        # create an instance for returned value
        returned_value = MyList([])
        
        if len1 == len2:
            
            # iterate over self and other
            for i in range(len1):
                # modify the returned value
                returned_value.list.append( self.list[i] + other.list[i] )
            return returned_value
            
        else:
            print("Different lengths! Please change your list.")
        
        
    def __str__(self):
        
        s = '['
        
        # iterate over all elements except for the last one
        for i in range( len(self.list)-1 ):
            
            s = s + repr(self.list[i]) + ','
         
        # add the last element and no comma.
        s = s + repr(self.list[-1]) + ']'
        
        return s      # s is a string. 
        
        
    def __repr__(self):
        
        
        s = 'My List is ['
        
        # iterate over all elements except for the last one
        for i in range( len(self.list)-1 ):
            
            s = s + repr(self.list[i]) + ','
         
        # add the last element and no comma.
        s = s + repr(self.list[-1]) + ']'
        
        return s
    
    def __eq__(self, other):
        """
        Corresponds to ==.
        """
        if self.list[0] == other.list[0]:
            return True
        else:
            return False
        
    def __iadd__(self, other):
        pass
````

#### Goal:
Simulate real-world financial operations with operator overloading and clear representations.

#### Task:
Design a class BankAccount that supports deposits, withdrawals, equality comparison, and account merging.

#### Requirements:

**Attributes:**
- owner: account holder’s name
- balance: float

**Methods:**
- deposit(self, amount) and withdraw(self, amount)

**Magic Methods:**
- \_\_add\_\_(self, other) -> merge two accounts (sum balances, combine owners as a string)
- \_\_iadd\_\_(self, amount) -> deposit money via +=
- \_\_eq\_\_(self, other) -> two accounts are equal if balances are equal
- \_\_str\_\_(self) -> human-readable summary, e.g.  
```Account(Chunyang): Balance = $1,200.00```
- \_\_repr\_\_(self) -> full technical representation, e.g. 
```BankAccount(owner='Chunyang', balance=1200)```

````python
class BankAccount:
    
    def __init__(self, owner, balance=0.0):
        self.owner = owner
        self.balance = balance
        
    def deposit(self, amount):
        self.balance = self.balance + amount
        
    def withdraw(self, amount):
        self.balance = self.balance - amount
        
    def __str__(self):
        summary = 'Account(' + self.owner + ') Balance = $' + str(self.balance) + ')'
        return summary
    
    def __repr__(self):
        summary = 'BankAccount(owner=' + repr(self.owner) + ', balance=' + repr(self.balance) + ")"
        return summary
````

## Inheritance

Goal: 1) Understand how inheritance works. 2) Understand super().

Motivating example: Rectangular and square

````python
class Rectangular:
    
    def __init__(self, length, width):
        self.var1  = length
        self.var2 = width
        
    def area(self):
        return self.var1 * self.var2
    
    def perimeter(self):
        return 2 * (self.var1 + self.var2)
    
    
class Square:
    
    def __init__(self, length):
        
        self.var = length
        
    def area(self):
        return self.var * self.var
    
    def perimeter(self):
        return 4 * self.var
````

Square and rectangular are related: square is a special type if rectangular. 

The code, however, does not reflect this.

By using inheritance, we are able to simplify the code. 

````python
class Rectangular:
    
    def __init__(self, length, width):
        self.var1  = length
        self.var2 = width
        
    def area(self):
        return self.var1 * self.var2
    
    def perimeter(self):
        return 2 * (self.var1 + self.var2)
    
    
class Square(Rectangular):
    
    def __init__(self, length):
        
        super().__init__(length, length)    # <==> Rectangular.__init__(length, length)
    
````

## Two concepts: Iterable and Iterator

- Iterable: an object that we can iterate over, e.g. list, set, and dictionary. There is an __iter__ magic method.

- Iterator: an object after applying __iter__ magic method. There is a __next__ magic method.


````python
L = [1,2,3,4]      # L is a list, and hence it is an iterable.
M = iter(L)        # after applying iter magic method, it becomes to an iterator.

# since M is an iterator, then we can apply next method.
print(next(M))
print(next(M))
````

````python
# Int, however, is not an iterable. Therefore, it does not have __iter__ magic method.
a = 10
iter(a)
````

````python
# How does for loop work in theory?

# In practice, we use the following code:
for i in L:
    print(i)
    
    
# In theory, it has two steps
# Step 1: applying __iter__ magic method to create an iterator

L = [1,2,3,4]
m = iter(L)      # an iterator

# Step 2: call __next__ magic method
print(next(m))
print(next(m))
print(next(m))
print(next(m))
print(next(m))
############### !!!!! We never do steps 1 and 2 in practice. But it is good to know this !!!
````

````python
# write my own for loop as a function (use try-except statement to handle StopIteration)

def my_for(iterable):

  # step 1: create an iterator
  iterator = iter(iterable)

  # step 2: call next method (use try-except statement to handle StopIteration)
  while True:
    try:
        el = next(iterator)
        print(el)
    except StopIteration:
        break

# test my code: it is equivalent to

### for i in L:
###     print(i)

L = [1,2,3,4]
my_for(L)
````

Before this cell, I tried to explain how __iter__ and __next__ work for built-in Python objects. 

Now, it is time to define a class and make it iterable.

````python
# Writing example 1: 

class Triple:

    def __init__(self, a, b, c):
        """
        a,b,c: numbers. 
        """
        self.a = a
        self.b = b
        self.c = c

    def __iter__(self):
        return TripleIterator(self)


class TripleIterator:

      def __init__(self, Triple):

        # self: it is an TripleIterator
        # Triple: an insatnce from Triple class.
    
        # I should put self.i in the __init__ because I need to reset my index
        self.i = -1       # please treat i as index. 

        # I am assigning attributes in Triple to self
        # In the future, I can use self.a, self.b, self.c. 
        # I do not need to pass Triple to __next__ method.
        self.a = Triple.a
        self.b = Triple.b
        self.c = Triple.c

    def __next__(self):
    
        self.i += 1       # change my index by adding 1 when I call next method

        if self.i == 0:
            return self.a
        if self.i == 1:
            return self.b 
        if self.i == 2:
            return self.c
        if self.i == 3:
            raise StopIteration
````

````python
ins = Triple(1,2,3)
print(ins.a, ins.b, ins.c)

for i in ins:
    print(i)
````

````python
# Writing example 2: I do not define Iterator class explicitly.

class Triple:

    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def __iter__(self):
        # set my index
        self.i = -1
        return self

    def __next__(self):

        self.i += 1

        if self.i == 0:
            return self.a
        if self.i == 1:
            return self.b 
        if self.i == 2:
            return self.c
        if self.i == 3:
            raise StopIteration
````

## Generator

**Motivation:** Using __iter__ and __next__ magic methods is complicated when we write the code. Generator provides a simple way to do that.

#### What is an generator?

- A generator function is a compact way to define an iterator without having to define a separated class.
- A generator function looks like a function that yields object rather than returning them.
- Calling a generator function doesn't immediately execute the code it contains
- Using next method to show the yield objects.

````python
def gen():

    print('Before 1')
    yield 1
    print('After 1')

    print('Before 2')
    yield 2
    print('After 2')

    print('Before 3')
    yield 3
    print('After 3')
````

````python
# it is a generator function, and calling generator function does not return the returned objects.
g = gen()
print(g)
````

````python
# calling next method to show the returned objects
print(next(g))
print(" ")
print(next(g))
print(" ")
print(next(g))
print(" ")
print(next(g))
````

````python
# Writing example 3: using generator to make class object iterable. The code is simple and short.
class Triple:

    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def __iter__(self):
        yield self.c
        yield self.b
        yield self.c
````

````python
ins = Triple(3,5,7)
for i in ins:
    print(i)
````

## Coding exercise: 

Define a class of integer that allows for loop. 

#### Goal: using while loop in the \_\_iter\_\_ or \_\_next\_\_ magic methods to enable iteration over a large amount of objetcs.

Method 1: We can have two classes, Int (iterable has __iter__ magic method) and IntIterator (iterator has __next__ magic method). I never use this one.

Method 2: I have one class containing \_\_iter\_\_ magic methods and \_\_next\_\_ magic methods.

Method 3: We can use generator. 


````python
L = [1,2,3,4]     # L is an iterable
print(iter(L))    # iter(L) is an iterator
````

````python
# Method 1
class Int:
    
    def __init__(self, num):
        self.num = num
        
    def __iter__(self):
        return IntIterator(self)
    
class IntIterator:
    
    def __init__(self, Int):
        self.i = -1
        self.num = Int.num
        
    def __next__(self):
        
        while self.i < self.num:
            self.i += 1
            return self.i
        if self.i >= self.num:
            raise StopIteration
````

````python
# Method 2:
class Int:
    
    def __init__(self, num):
        """
        Initialization.
        Args:
        num: python int object.
        """
        self.num = num
        
    def __iter__(self):
        """
        After applying iter, we have index for my instance.
        """
        self.i = 0      # my index
        return self
    
    def __next__(self):
        
        while self.i < self.num:
            self.i += 1            # change index
            return self.i          # return value
        if self.i >= self.num:     # check the end point
            raise StopIteration    # raise StopIteration when we stop
    
## My expectation:
a = Int(10)
for i in a:
    print(i)
    
# 1,2,3,...,10
````

````python
# Method 3:
class Int:
    
    def __init__(self, num):
        self.num = num
        
    def __iter__(self):
        self.i = 0
        
        while self.i<self.num:
            self.i += 1
            yield self.i  

# Test example:
a = Int(10)
for i in a:
    print(i)
````

The following example discusses how we can minic enumerate method for your class:

````python
# Recall enumerate
L = [1,2,3,4]
for idx, val in enumerate(L):
    print(idx, val)
    print(" ")
````

````python
# enumerate is not a magic method.
# enumerate can be use to all iterables.
# We apply enumerate to a user-defined class Int
class Int:
    
    def __init__(self, num):
        self.num = num
        
    def __iter__(self):
        self.i = 0
        
        while self.i<self.num:
            self.i += 1
            yield self.i  

# Test example:
a = Int(10)
for idx, val in enumerate(a):
    print(idx, val)
````

````python
# We do not call enumerate function
# But we can modify our code to have a similar result as calling enumerate function.

class Int:
    
    def __init__(self, num):
        self.num = num
        
    def __iter__(self):
        self.i = 0
        
        while self.i<self.num:
            self.i += 1
            yield self.i-1, self.i  

# Test example:
a = Int(10)
for idx, val in a:
    print(idx, val)
````
