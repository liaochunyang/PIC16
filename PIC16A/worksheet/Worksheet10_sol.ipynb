#### Step 11. Compute average PTS, average REB, average AST, average STL, and average BLK for each player and create a new dataFrame contains player name and computed statistics

average points = PTS/Game Played

Column names: ['name','PTS','REB','AST','STL','BLK']

```python
new_df = pd.DataFrame(df['Name'])
new_df['PTS'] = df['PTS']/df['Game Played']
new_df['REB'] = df['REB']/df['Game Played']
new_df['AST'] = df['AST']/df['Game Played']
new_df['STL'] = df['STL']/df['Game Played']
new_df['BLK'] = df['BLK']/df['Game Played']
new_df
```

#### Step 12. Select players who play either less than 300 minutes or less than 20 games, and then return their name, average PTS and average REB

```python
condition = (df['Minutes']<300) | (df['Game Played']<20)
new_df[condition].loc[:,['Name','PTS','REB']]
```
