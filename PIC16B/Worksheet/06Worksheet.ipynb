# Worksheet: Optimization

The goal of this exercise is implementing GD algorithm to solve the following optimization problem
$$ \mathop{\mathbf{minimize}}_{x,y} \quad f(x,y) = \exp(x+3y-0.1) + \exp(x-3y-0.1) + \exp(-x-0.1) $$

#### Gradient

As we discussed in the lecture, implementing GD algorithm requires the gradient of the objective function. Please compute the gradient by hand. You are welcome to ask your TA for help.

**Remark: It is possible to compute the gradient of the objective function by-hand in this worksheet. However, it is impossible to compute by hand if the objective function is too complicated. In that situation, we can use auto-differentiation to compute the gradient numerically.**

#### Solve optimization problem by hand

Setting the gradient to be zero gives two equations. Solving the equations gives you the desired point $(x^*,y^*)=(\frac{1}{2}\log(\frac{1}{2}), 0)$ which minimizes the objective function. In other words, we know that the minimizer is $(x^*,y^*)=(\frac{1}{2}\log(\frac{1}{2}), 0)$. The minimizer is given to verify the correctness of your GD algorithm. You are not allowed to use it when you write your code.

**Remark: This optimization problem is not too complicated and we are able to solve it by-hand. However, it is impossible to find minimizers by-hand if the optimization problem is hard.**
