## Part 2: Shape of the output

Run the following examples and explain why the outputs have different shapes even if the calculations are the same.

```python
n = 2
A = np.random.randint(1, 10, size = (n, n))
x = np.array([1,1])
x1 = np.array([[1],[1]])
y = A@x
y1 = A@x1

print('Matrix-vector multiplication is implemented:')
print(y)
print(y1)

print('Shapes are different:')
print(y.shape, y1.shape)
```
