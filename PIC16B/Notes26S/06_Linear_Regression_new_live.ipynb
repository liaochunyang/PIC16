## Supervised and Unsupervised learning

Supervised learning means that you can use both input $x$ and output $y$ to train a model. The output $y$ is also called *labels*.

If the labels are "continuous" real numbers (e.g. stock price, house price, insurance fee, height, weights, and etc ), then it is **regression problem**.

If the labels are discrete categories (e.g. dog vs cat, True vs False, handwritten digits numbers(0-9), 0 vs 1, or -1 vs 1, and etc), then it is **classification problem**.


We will also discuss unsupervised learning problem where labels $y$ is not available. Two typical unsupervised learning problems are **clustering** and **dimension reduction**.

# Encoding:

Encoding means transforming categorical features to numerical features.

- One-hot encoding: each category is represented by a binary/sparse vector, and each category becomes to a new column.

- Label encoding: integer representation. Each category is represented by an integer. 

```python
# Two functions to do encodings.
def one_hot(df, cols):
    """
    df: pandas DataFrame
    param: cols a list of columns to encode 
    return a DataFrame with one-hot encoding
    """
    for each in cols:
        dummies = pd.get_dummies(df[each], prefix=each, drop_first=False)
        df = pd.concat([df, dummies], axis=1)
    return df


def label(df, cols):
    
    df_new = df.copy()
    
    for each in cols:
        
        label_encoder = LabelEncoder()
        #dummies = label_encoder.fit_transform(df[each])
        #dummies = pd.DataFrame(dummies, columns=[each])
        #df = pd.concat([df, dummies], axis=1)
        
        label_encoder = LabelEncoder()
        df_new[each] = label_encoder.fit_transform(df_new[each])
        
    return df_new
```

# Linear Regression model

Linear regression model describes the relationship between features and targets using straight line (2D) or linear hyperplane (high-dimensional), see the following picture:

![Linear_least_squares_example2.svg](https://community.cloudera.com/t5/image/serverpage/image-id/25068iFF075A5AEC3B8528/image-size/medium?v=v2&px=400)


**Fact: A straight line can be modeled as y=kx+b where k is called slope and b is called intercept.**

Our goal is to find $k$ and $b$ such that the obtained straight line fits your data well (how to define this mathematically?)


**Formula:**

We set up the following minimization problem

$$\mathop{\mathbf{minimize}}_{k,b} \frac{1}{m}\sum_{i=1}^{m} (kx^{(i)} + b - y_i)^2, $$
where m is the number of data samples.

**High-dimensional case:**

In many real world applications, you are asked to use more than one features to predict the output, see example below.

In high dimension case, the linear regression model can be written as $w_0 + w_1x_1 + ... + w_dx_d$, where $d$ is the dimension of the inputs data $x$, i.e. $x=(x_1, x_2, ..., x_d)\in\mathbb{R}^d$. Our goal is to find coefficients $w_0, ..., w_d$.

Given training samples $\{x^{(i)}, y^{(i)}\}_{i=1}^m$, finding coefficients is done by solving the following optimization problem

$$ \frac{1}{m} \sum_{i=1}^m (w_0 + w_1x^{(i)}_1 + ... + w_dx^{(i)}_d - y^{(i)})^2. $$

**Training machine learning models:**

Usually, training machine learning models is equivalent to solving an optimization problem.

#### Some mathematics

**Mean-squared error (MSE)**: $\frac{1}{m}\sum_{i=1}^m (\hat{y}_i-y_i)^2$, where 
- $\hat{y}_i$ is the i-th predicted value by your model. 
- $y_i$ is the i-th true value from your dataset.
- It computes training MSE if y_i are training outputs and $\hat{y}$ are computed by training inputs.
- We can also compute test MSE when we use test samples.

When we compare models, we use **Test MSE** because we want to measure the model performance on the unseen dataset.

Test mean-squared error: good for comparing models, no interpretation. When we choose models, we prefer small mean-squared error.

**Relative error** measures how large your mistake is as a fraction of the true value, the math formula is
$$ \frac{error}{True Value} $$

For multiple test samples, we define the **average relative test error** as
$$ \frac{1}{m}\sum_{i=1}^m \frac{|y_i-\hat{y}_i| }{|y_i|}$$

# Regularization

Regualization was proposed to reduce model complexity. It prevents **overfitting** by adding a penalty term to the cost function, discouraging overly complex models.

**Overfitting:** the model has perfect performance on training samples, but fails to generalize to any unseen samples.

In the classical machine learning theory, people showed that overfitting is harmful. In recent years, people realize that overfitting can be benign under some certain conditions. Understanding the benign overfitting mechanism is an active research area. 


## Ridge Regression and Least Absolute Shrinkage and Selection Operator (LASSO)


Please see this review for the details:[Lasso and Ridge regression: a comprehensive review of applications
and developments in machine learning](https://link.springer.com/article/10.1007/s41060-025-00957-y)

#### Ridge Regression:

Ridge Regression solves the following optimization problem:

$$ \frac{1}{m} \sum_{i=1}^m (w_0 + w_1x^{(i)}_1 + ... + w_dx^{(i)}_d - y^{(i)})^2 + \lambda \sum_{j=1}^d w_d^2,$$

where $\lambda$ is a regularization parameter. 

#### LASSO:

LASSO solves the following optimization problem:


$$ \frac{1}{m} \sum_{i=1}^m (w_0 + w_1x^{(i)}_1 + ... + w_dx^{(i)}_d - y^{(i)})^2 + \lambda \sum_{j=1}^d |w_d|,$$

where $\lambda$ is a regularization parameter. 

#### Difference:

LASSO can produce sparse coefficients.

#### The role of regularization parameter:

Larger $\lambda$ means stronger regularization, since it focuses more on the regularization term (the second term). When $\lambda\to0$, the training algorithm forces the model to memorize the traning samples, which leads to overfitting.

## Repeat Experiments to remove randomness

Notice that train-test split will create training and testing samples randomly. Different training samples may results in different models, and hence we should repeat our experiments to remove the effect of randomness.

```python
def compute_test_error(X, y):
    """
    Compute test error. 
    Args:
    X: numpy array containing all features.
    y: target. A numpy array.
    Returns:
    mse: mean squared error.
    rel_se: relative squared error.
    """
    # train test split: each time you will randomly select training and test samples.
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
    
    # model initialization
    model = LinearRegression()
    # trianing this model
    model.fit(X_train, y_train)
    
    # prediction 
    y_pred = model.predict(X_test)
    
    # compute the relative squared error
    rel = np.mean( np.abs(y_test-y_pred)/np.abs(y_test) )
    
    return rel
```
