## Basic objects: Series and DataFrame

#### Series: Just one column

#### DataFrame: Multiple columns


```python
# Create a pandas seires
df = pd.Series([10, 'A', 9.08])
df
```

```python
print(df)
```

```python
# create a pandas DataFrame
d = {'student name': ['Tom', 'Jerry', 'Smith', 'Bob'], 'UID':[100,101,102,103]}
df = pd.DataFrame(d)

# print the dictionary
print(df)
print(" ")

# 'student name' and UID are column names
print(df.columns)
print(" ")

# labels (my terminology), it is called index in pandas
print(df.index)
```

```python
# create a pandas DataFrame
labels = ['Tom', 'Jerry', 'Smith', 'Bob']
d = {'UID':[100,101,102,103]}

df = pd.DataFrame(d, index=labels)

# print the dictionary
print(df)
print(" ")

# 'student name' and UID are column names
print(df.columns)
print(" ")

# labels (my terminology), but it is called index in pandas
print(df.index)
```

```python
# Numpy array to a dataframe
labels = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
a = np.random.uniform(-1,1,size=(10,4))
df = pd.DataFrame(a,index=labels)
df
```

```python
# It is from dataframe to numpy array
b = df.to_numpy()
print(b)
print(type(b))
```

#### Select by rows

1. `pd.loc`: selection by using labels. 
2. `pd.iloc`: selection by using index.

In Python, when we do selection/slicing using index, the end point is excluded, e.g. range(5), L=[1,2,3,4] L[0:2].

```python
A = {'A': ['True','True','False','False'], 'B':[8,7,6,5] }
labels = [100,101,102,103]
df_test = pd.DataFrame(A, index=labels)
df_test
```

```python
# loc: 102 is included because it is labels (not index).
df_test.loc[100:102]
```

```python
# iloc: select the rows with index 0, 1, and 2., so 3 is excluded because it is index. 
df_test.iloc[0:3]
```

#### selection by condition



```python
# # of children <= 1
condition = df['children'] <= 1
df[condition]
```

```python
# multiple condition: age > 30 and region="southwest"
condition = (df['age'] > 30) & (df['region']=="southwest")
df[condition]
```

```python
# conditionL age > 30 or region="southwest"
condition = (df['age'] > 30) | (df['region']=="southwest")
df[condition]
```

## Data Transformation

```python
# I want to have a new column (Y) with normalized charges
df['Y'] = df['charges'] / 100
df
```

```python
# I want to update column 'bmi' with normalized bmi
df['bmi'] = df['bmi'] / 30
df
```

```python
# If the operation is complicated, 
# we can define a function and use `pd.apply` to do entry-wise transformation.

def fun(x):
    
    return (round(x) + 10) / 5.76

df['bmi'].apply(fun) 
```
