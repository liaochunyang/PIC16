# Exercise 1

The kth triangular number (https://en.wikipedia.org/wiki/Triangular_number) is the sum of natural numbers up to and including k. Write a function named **get_triangular_numbers** that returns a list of the first k triangular numbers.
For example, the sixth triangular number is 1+2+3+4+5+6=21, and running get_triangular_numbers with an argument of k=6 should output [1, 3, 6, 10, 15, 21].


```python
# Your function goes here
def get_triangular_numbers(k):
    
    M = []
    summation = 0
    for i in range(1,k+1):
        summation = summation + i
        M.append(summation)
    
    return M
```

```python
# Test example:
k = 3
print(get_triangular_numbers(k))
# should return [1，3，6]
```

# Exercise 2: 

The function **get_consonants** taks a string `s` as an input, and returns a list of the letters in `s` except for vowels, spaces, commas, and periods. For the purposes of this example, an English vowel is any of the letters ["a", "e", "i", "o", "u"]. 

```python
# Your solution goes here
def get_consonants(s):
    
    L = []
    
    for i in s:
        if i not in ["a", "e", "i", "o", "u", ' ', ',']:
            L.append(i)
            
    return L
```

```python
# Test example
s = "make it so, number one"
print(get_consonants(s))
# should print ["m", "k", "t", "s", "n", "m", "b", "r", "n"] or ['m', 'k', 't', 's', 'n', 'm', 'b', 'r', 'n']
```

# Exercise 3:

(a) Compute the sumber of first N integers, i.e. sum of numbers 1-N. The function name is **addition** and takes an integer N as input.

(b) A **recursive function** is a function that calls itself again and again. Use recursive function to compute the sum of first N integers. This function is named **recursive_addition**.


```python
# Your solution goes here
def addition(n):
    
    s = 0
    for i in range(1,n+1):
        s = s+i
    return s

def recursive_addition(m):
    
    if m == 1:
        return m
    else: 
        return m + recursive_addition(m-1)
```

```python
# Test example 
print(addition(10))              # should print 55
print(recursive_addition(20))    # should print 210
```

# Exercise 4

(a) The function `get_list_of_powers` takes in a list `X` and integer `k` and returns a list `L` whose elements are themselves lists. The `i`th element of `L` contains the powers of `X[i]` from `0` to `k`. 

For example, running `get_list_of_powers` with inputs `X = [5, 6, 7]` and `k = 2` will return `[[1, 5, 25], [1, 6, 36], [1, 7, 49]]`.

```python
# Your solution here
def get_list_of_powers(X,k):
    
    L = [[x**j for j in range(k+1)] for x in X ]
    
    return L
```

```python
get_list_of_powers(X=[5,6,7],k=2)  # should return [[1, 5, 25], [1, 6, 36], [1, 7, 49]]
```
